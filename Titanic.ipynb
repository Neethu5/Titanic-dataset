```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import svm
data = pd.read_csv('C:/Users/ayush/Desktop/Kaggle/train.csv')
data.head(3)
```

```python
y_target = data['Survived']
# Y_target.reshape(len(Y_target),1)
x_train = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
x_train.head()
```

```python
fare = pd.DataFrame(x_train['Fare'])
min_max_scaler = preprocessing.MinMaxScaler()
newfare = min_max_scaler.fit_transform(fare)
x_train['Fare'] = newfare
x_train
```

```python
x_train.isnull().sum()
```

```python
# Fill the NAN values with the median values in the datasets
x_train['Age'] = x_train['Age'].fillna(x_train['Age'].median())
print("Number of NULL values" , x_train['Age'].isnull().sum())
print(x_train.head(3))
```

```python
x_train['Sex'] = x_train['Sex'].replace('male', 0)
x_train['Sex'] = x_train['Sex'].replace('female', 1)
print(x_train.head())
```

```python
# Dividing the data into train and test data set
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_target, test_size = 0.4)
```

```python
clf = svm.SVC(kernel='poly', verbose=True)

clf.fit(X_train, Y_train)
```

```python
print(clf.predict(X_test))
print(clf.score(X_test, Y_test))
```

```python
## Testing the model.
test_data = pd.read_csv('C:/Users/ayush/Desktop/Kaggle/test.csv')
test_data.head(3)
```

```python
### Preprocessing on the test data
test_data = test_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]

fare = pd.DataFrame(test_data['Fare'])
min_max_scaler = preprocessing.MinMaxScaler()
newfare = min_max_scaler.fit_transform(fare)
test_data['Fare'] = newfare
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].median())
test_data['Sex'] = test_data['Sex'].replace('male', 0)
test_data['Sex'] = test_data['Sex'].replace('female', 1)
print(test_data)
# x_train.isnull().sum()
```
